## Problem Statement

- Tic-Tac-Toe is very popular and is quite simple in itself. It is a two-player game. In this game, there is a board with 3×3 squares. You will have to use the popular Pygame library for rendering graphics on a display window.


- In this game a player can choose between two symbols with his opponent, the usual games use “X” and “O”. If the first player chooses “X”, then the second player must play with “O” and vice versa.


- A player marks one of the 3×3 squares with his symbol (perhaps “X” or “O”) and he aims to create a straight line horizontally or vertically or diagonally with two intensions:


- Create a straight line before your opponent to win the game.


- Prevent his opponent from creating a straight line first.


- If no one can logically create a straight line with its symbol, the game ends in a tie. So there are only three possible outcomes: one player wins, his opponent (human or computer) wins, or there is a tie. You will prepare this Tic Tac Toe GUI with Python for two players.

## Tasks to be performed:

- Use Pygame library to build Tic-Tac-Toe game in python.

- Create the display window for our game.

- Draw the grid on the canvas where you will play Tic-Tac-Toe.

- Draw the status bar below the canvas to show which player’s turn is it and who wins the game.

- When someone wins the game or the game is a draw then reset the game.

- You need to run our game inside an infinite loop. It will continuously look for events and when a user presses the mouse button on the grid we will first get the X and Y coordinates of the mouse. Then we will check which square the user has clicked. Then we will draw the appropriate ‘X’ or ‘O’ image on the canvas. 


```python
# Initailizing variables

t_var = 'x'                      # For storing the 'X' or 'O' value as character.

game_winner = None               # To store the winner's value at any instant of game, intially no one is winner.
  
game_draw = None                 # To check if the game is a draw, intially game is not draw.

width = 400                      # To set width of the game window.
  
height = 400                     # To set height of the game window.
  
background = (255, 255, 255)     # To set background color of the game window.
  
line_color = (0, 0, 0)           # color of the straightlines on that white game board, dividing board into 9 parts.

#TicTacToe 3x3 board.
board = [[None]*3, [None]*3, [None]*3]        # design of a 3 * 3 empty board in canvas.
  
pg.init()                        # initializing the pygame window.
  
fps = 30                         # setting fps (#30 frames per sec).

CLOCK = pg.time.Clock()          # To track time using clock.
  
screen = pg.display.set_mode((width, height + 100), 0, 32)   # To build the screen for display of game. 
  
pg.display.set_caption("Tic-Tac-Toe_Game")          # setting up a nametag for the game window
   
# Loading images
opening = pg.image.load(opening)           # loading image of opening screen
x_img = pg.image.load(x_img)               # loading image of X symbol
o_img = pg.image.load(o_img)               # loading image of O symbol
ending = pg.image.load(closing)            # loading image of closing screen

# Resizing images
x_img = pg.transform.scale(x_img, (80,80))
o_img = pg.transform.scale(o_img, (80,80))
opening = pg.transform.scale(opening, (width, height+100))
closing = pg.transform.scale(ending, (width, height+100))


# Function for displaying game screen
def game_opening():
    screen.blit(opening,(0,0))       # Welcome screen.
    pg.display.update()
    time.sleep(1)
    screen.fill(background)
    
    # Drawing vertical lines
    pg.draw.line(screen,line_color,(width/3,0),(width/3, height),7)
    pg.draw.line(screen,line_color,(width/3*2,0),(width/3*2, height),7)
    
    # Drawing horizontal lines
    pg.draw.line(screen,line_color,(0,height/3),(width, height/3),7)
    pg.draw.line(screen,line_color,(0,height/3*2),(width, height/3*2),7)
    game_status()
    

# Function for status of game: winner, draw or player's turn.
# Everytime a message is displayed depending on whether its player's turn, or declaring winner or game is draw.
def game_status():
    global game_draw

    if game_winner is None:
        message = t_var.upper() + "'s Turn"
    else:
        message = game_winner.upper() + " has won!"
    if game_draw:
        message = 'Game Draw!'

    font = pg.font.Font(None, 30)
    text = font.render(message, 1, (255, 255, 255))

    # displaying message on the game board
    screen.fill ((0, 0, 0), (0, 400, 500, 100))
    text_rect = text.get_rect(center=(width/2, 450))
    screen.blit(text, text_rect)
    pg.display.update()


# Function to check if a player has won or the game is graw.
# A player wins if he is able to draw vertical or horizontal or diagonal lines using the symbols.
def check_winner():
    global board, game_winner, game_draw

    # check if player wins by drawing horizontal line passing though any rows
    for row in range (0,3):
        if (( board [row][0] == board[row][1] == board[row][2]) and( board[row][0] is not None)):
            # this row won
            game_winner = board[row][0]
            # to draw winning line
            pg.draw.line(screen, (250,0,0), (0, (row + 1)*height/3 -height/6),(width, (row + 1)*height/3 - height/6 ), 4)
            break

    # check if player wins by drawing vertical line passing though any columns
    for col in range (0, 3):
        if ( board[0][col] == board[1][col] == board[2][col]) and ( board[0][col] is not None):
            # this column won
            game_winner = board[0][col]
            # to draw winning line
            pg.draw.line(screen, (250,0,0),((col + 1)* width/3 - width/6, 0),((col + 1)* width/3 - width/6, height), 4)
            break

    # check if player wins by drawing diagonal line 
    if ( board[0][0] == board[1][1] == board[2][2]) and ( board[0][0] is not None):
        # game won diagonally left to right i.e. \
        game_winner = board[0][0]
        # to draw winning line
        pg.draw.line (screen, (250,70,70), (50, 50), (350, 350), 4)
       
    # check if player wins by drawing diagonal line 
    if ( board[0][2] == board[1][1] == board[2][0]) and ( board[0][2] is not None):
        # game won diagonally right to left i.e. /
        game_winner = board[0][2]
        # to draw winning line
        pg.draw.line (screen, (250,70,70), (350, 50), (50, 350), 4)
    
    # if no line could be drawn using any symbols, then game is draw.
    if(all([all(row) for row in board]) and game_winner is None ):
        game_draw = True
    
    # Calling function to check status of game.
    game_status()

# Function to display player's turn depending on current scenario
def drawBoard(row,col):
    global board,t_var
    if row==1:
        posx = 30
    if row==2:
        posx = width/3 + 30
    if row==3:
        posx = 2*width/3 + 30

    if col==1:
        posy = 30
    if col==2:
        posy = height/3 + 30
    if col==3:
        posy = 2*height/3 + 30
    board[row-1][col-1] = t_var
    if(t_var == 'x'):
        screen.blit(x_img,(posy,posx))
        t_var = 'o'
    else:
        screen.blit(o_img,(posy,posx))
        t_var = 'x'
    pg.display.update()
    

# Function to get the coordinates of clicked area.
def mouseClick():
    #get coordinates of mouse clicked
    x,y = pg.mouse.get_pos()

    #get column of mouse clicked
    if(x<width/3):
        col = 1
    elif (x<width/3*2):
        col = 2
    elif(x<width):
        col = 3
    else:
        col = None
        
    #get row of mouse click 
    if(y<height/3):
        row = 1
    elif (y<height/3*2):
        row = 2
    elif(y<height):
        row = 3
    else:
        row = None

    if(row and col and board[row-1][col-1] is None):
        global t_var
        
        #draw the x or o on screen
        drawBoard(row,col)
        # check status of game player
        check_winner()
        
# Function to reset game after game is over.
def reset_game():
    global board, game_winner, t_var, game_draw
    time.sleep(3)
    t_var = 'x'
    game_draw = False
    game_winner=None
    board = [[None]*3,[None]*3,[None]*3]
    
    # Calling function to start game again.
    game_opening()
    

# Calling Game opening function for the first time.
game_opening()

# To run the game loop infinte times.
while True:
    for event in pg.event.get():
        # if player quits or closes the game window.
        if event.type == QUIT:
            pg.quit()
            sys.exit()           # system exits, all variables are destroyed.
            
        # else if user clicked then restart game again.
        elif event.type == MOUSEBUTTONDOWN:
            # the player clicked; place an X or O
            mouseClick()
            if(game_winner or game_draw):
                time.sleep(1)
                screen.blit(closing,(0,0))       # to show game over screen
                pg.display.update()
                reset_game()                     # to reset game.
            
    pg.display.update()
    CLOCK.tick(fps)
```
